### Heart attack prediction

```python
import pandas as pd
path="D:\\Internship_project_Dataset\\Heart attack prediction\\heart.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
inp=data.drop('output','columns')
out=data['output']
print('..INPUT DATA..')
print(inp,'\n')
print("..OUTPUT DATA..")
print(out,'\n')
import sklearn
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(inp,out,train_size=0.8)
print('X_train')
print(x_tr,'\n')
print("X_test")
print(x_te,'\n')
print('Y_train')
print(y_tr,'\n')
print('y_test')
print(y_te,'\n')

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_tr,y_tr)
res=model.predict(x_te)
print('...Predicted_Result...')
print(res,'\n')
print('...Y_test Comparision...')
print(y_te,'\n')
from sklearn.metrics import accuracy_score
acc=accuracy_score(res,y_te)*100
print("Accuracy:",str(acc),'\n')
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(res,y_te)
print('...Using confusion matrix for Accuracy_Prediction...')
print(cm)
```

### Stroke Prediction

```python
import pandas as pd
path="D:\\Internship_project_Dataset\\Stroke Predection\\healthcare-dataset-stroke-data.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
data=data.fillna(method='ffill')
print(data)

import sklearn
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data.gender=le.fit_transform(data.gender)
data.smoking_status=le.fit_transform(data.smoking_status)
data.ever_married=le.fit_transform(data.ever_married)
data.work_type=le.fit_transform(data.work_type)
data.Residence_type=le.fit_transform(data.Residence_type)
print(data)
print(data.info(),'\n')                       
#separating input and output
inp=data.drop(['stroke','id'],'columns')
out=data['stroke']
print("...INPUT DATA...")
print(inp,'\n')
print('...OUTPUT DATA...')
print(out,'\n')

#import sklearn
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te=train_test_split(inp,out,train_size=0.8)
print('...X_train...')
print(x_tr,'\n')
print('...X_test...')
print(x_te,'\n')
print('...Y_train...')
print(y_tr,'\n')
print('...Y_test...')
print(y_te,'\n')

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_tr,y_tr)
res=model.predict(x_te)
print('...Predicted_Result...')
print(res,'\n')
print('...Y_test Comparision...')
print(y_te)
from sklearn.metrics import accuracy_score
acc=accuracy_score(res,y_te)*100
print('Predicted_Accuracy:',acc)
```
